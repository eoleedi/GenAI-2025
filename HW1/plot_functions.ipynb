1. 簡單的cos(x)函數

```python
# prompt: draw y=sin(x)

x = np.linspace(0, 10, 100)
y = np.cos(x)
plt.plot(x,y)
plt.show()
```

2. 一元二次方程式

```python
# prompt: draw 一元二次方程式

x = np.linspace(-5, 5, 100)
y = x**2 + 2*x + 1
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("一元二次方程式: y = x^2 + 2x + 1")
plt.grid(True)
plt.show()
```

```python
# prompt: 用函數話一個愛心

import matplotlib.pyplot as plt
import numpy as np

def heart_function(t):
  x = 16 * np.sin(t)**3
  y = 13 * np.cos(t) - 5 * np.cos(2*t) - 2 * np.cos(3*t) - np.cos(4*t)
  return x, y

t = np.linspace(0, 2*np.pi, 1000)
x, y = heart_function(t)

plt.figure(figsize=(6,6))
plt.plot(x, y, color='red')
plt.title('愛心函數', fontsize=16)
plt.show()
```
