### 4. 用 Gradio 打造你的對話機器人 Web App!

```python
import gradio as gr
```

```python
messages = [{"role":"system",
             "content":system}]
```

```python

def mychatbot(prompt, chat_history=[]):
    chat_history.append({"role": "user", "content": prompt})
    chat_completion = client.chat.completions.create(
        messages=chat_history,
        model=model,
    )
    reply = chat_completion.choices[0].message.content
    chat_history.append({"role": "assistant", "content": reply})

    # Format chat history for display
    chat_messages = []  # changed from chat_message
    for message in chat_history:
        if message["role"] == "user":
            chat_messages.append(f"User: {message['content']}")
        else:
            chat_messages.append(f"Bot: {message['content']}")
    return chat_messages, chat_history  # 返回空字符串和更新的聊天記錄
```

```python
iface = gr.Interface(
    mychatbot,
    inputs=["text",gr.State([])],  # 添加 state 輸入以保存聊天記錄
    outputs=["list", "state"],  # 添加 state 輸出以更新聊天記錄
    title=title,
    description=description,
)
```

```python
iface.launch(share=True, debug=True)
```
